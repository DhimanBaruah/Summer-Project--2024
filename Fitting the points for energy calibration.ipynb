```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.colors import LogNorm
import seaborn as sns
from scipy.optimize import curve_fit
from scipy import stats
from scipy.special import erf
```

```python
ADC_channel = np.array([  447.56066337 , 1473.49098918 ,  644.17417362])
energy = np.array([           0.511       , 1.2745             ,    0.66166        ]) #in MeV   1473.4947349473496+-2.52160183138928 , 447.5624756247562+-2.563900403065928 , 644.17417362+-2.80351526651775   
plt.scatter( ADC_channel , energy , color = 'black')
res = stats.linregress( ADC_channel , energy  )
x = np.linspace( 0 , max(ADC_channel)+ 100)
plt.plot(x , res.intercept + res.slope*( x ), 'r', label='fitted line')
plt.legend()
plt.xlabel("ADC")
plt.ylabel("Energy in MeV")
plt.savefig("linear_fit.png")
plt.show()
```
